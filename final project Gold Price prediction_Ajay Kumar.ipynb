```python
## Gold Price Prediction Using Time Series Forecasting ## 

# 1. Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping

plt.rcParams["figure.figsize"] = (12,6)

# 2. Load dataset
file_path = r"C:\Users\Ajay\Desktop\Gold price prediction\Gold Price (2013-2023).csv"
df = pd.read_csv(file_path)

print("Dataset preview:")
print(df.head())
print("Columns:", list(df.columns))

# 3. Process date column
date_col = [c for c in df.columns if "date" in c.lower()][0]
df[date_col] = pd.to_datetime(df[date_col])
df.set_index(date_col, inplace=True)
df.sort_index(inplace=True)

# 4. Process price column
price_col = None
for col in df.columns:
    if any(x in col.lower() for x in ["price", "close", "gold"]):
        price_col = col
        break
if price_col is None:
    price_col = df.columns[0]

print("Using price column:", price_col)

df[price_col] = df[price_col].astype(str)
df[price_col] = df[price_col].str.replace(",", "", regex=True)
df[price_col] = df[price_col].str.replace("$", "", regex=True)
df[price_col] = df[price_col].str.replace("", "", regex=True)
df[price_col] = pd.to_numeric(df[price_col], errors="coerce")
df[price_col].fillna(method="ffill", inplace=True)
df.dropna(inplace=True)

# 5. Exploratory Data Analysis
plt.plot(df.index, df[price_col], label="Gold Price")
plt.title("Gold Price Trend (20132023)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# 6. Stationarity test
adf_result = adfuller(df[price_col])
print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])
if adf_result[1] > 0.05:
    print("Data is non-stationary; differencing handled in ARIMA")
else:
    print("Data is stationary")

# 7. ARIMA model
arima_model = ARIMA(df[price_col], order=(5,1,0))
arima_fit = arima_model.fit()
arima_forecast = arima_fit.forecast(steps=30)

plt.plot(df[price_col], label="Actual Price")
plt.plot(arima_forecast, label="ARIMA Forecast", linestyle="--")
plt.title("ARIMA Gold Price Forecast")
plt.legend()
plt.show()

# 8. Prepare data for LSTM (with train-test split)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[[price_col]])

window_size = 60
train_size = int(len(scaled_data) * 0.8)

train_data = scaled_data[:train_size]
test_data = scaled_data[train_size - window_size:]

X_train, y_train = [], []
X_test, y_test = [], []

for i in range(window_size, len(train_data)):
    X_train.append(train_data[i-window_size:i, 0])
    y_train.append(train_data[i, 0])

for i in range(window_size, len(test_data)):
    X_test.append(test_data[i-window_size:i, 0])
    y_test.append(test_data[i, 0])

X_train = np.array(X_train).reshape(-1, window_size, 1)
y_train = np.array(y_train)

X_test = np.array(X_test).reshape(-1, window_size, 1)
y_test = np.array(y_test)

# 9. Build and train improved LSTM model
model = Sequential()
model.add(Bidirectional(LSTM(64, return_sequences=True), input_shape=(window_size,1)))
model.add(Dropout(0.2))
model.add(LSTM(64))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mean_squared_error")

early_stop = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)

model.fit(X_train, y_train, epochs=50, batch_size=32, 
          validation_data=(X_test, y_test), callbacks=[early_stop], verbose=1)

# 10. Evaluate on test data
predicted_test = model.predict(X_test)
predicted_test = scaler.inverse_transform(predicted_test)
actual_test = scaler.inverse_transform(y_test.reshape(-1,1))

rmse_test = np.sqrt(mean_squared_error(actual_test, predicted_test))
print("Test RMSE:", rmse_test)

plt.plot(actual_test, label="Actual Price")
plt.plot(predicted_test, label="Predicted Price")
plt.title("Improved LSTM: Actual vs Predicted Gold Price")
plt.legend()
plt.show()

# 11. Future 30-day forecasting
future_predictions = []
last_sequence = scaled_data[-window_size:]

for _ in range(30):
    next_val = model.predict(last_sequence.reshape(1,window_size,1))
    future_predictions.append(next_val[0,0])
    last_sequence = np.vstack([last_sequence[1:], next_val])

future_predictions = scaler.inverse_transform(
    np.array(future_predictions).reshape(-1,1)
)
future_dates = pd.date_range(start=df.index[-1] + pd.Timedelta(days=1), periods=30)

plt.plot(df.index[-120:], df[price_col].tail(120), label="Historical Prices")
plt.plot(future_dates, future_predictions, label="Future Forecast", linewidth=3)
plt.axvline(df.index[-1], linestyle="--", alpha=0.5)
plt.title("Gold Price Forecast for Next 30 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```
